```python
x = happydf.GDP_per_capita
y = happydf.Score

fig = mpl.figure(figsize = (14,7))
ax = fig.add_subplot()
ax.grid()

ax.set_xlabel("GDP")
ax.set_ylabel("Score")
ax.set_title("GDP vs Score")

ax.scatter(x,y)
```

```python
data_linreg = np.polyfit(happydf.GDP_per_capita, happydf.Score, 1)
print(data_linreg)
```

```python
x = happydf.GDP_per_capita
y = happydf.Score

xp = np.linspace(-0.02, 1.75, 156)
lr = np.polyval(data_linreg, xp)

fig = mpl.figure(figsize = (14,7))
ax = fig.add_subplot()
ax.grid()

ax.set_xlabel("GDP")
ax.set_ylabel("Score")
ax.set_title("GDP vs Score")

ax.scatter(x,y, color = "lightgrey")
ax.plot(xp, lr, color = "blue")
```

```python
x = happydf.Healthy_life_expectancy
y = happydf.Score

fig = mpl.figure(figsize = (14,7))
ax = fig.add_subplot()
ax.grid()

ax.set_xlabel("Healthy_life_expectancy")
ax.set_ylabel("Score")
ax.set_title("Life expecensy vs Score")

ax.scatter(x,y)
```

```python
data_polreg = np.polyfit(happydf.Healthy_life_expectancy, happydf.Score, 2)
print(data_polreg)
```

```python
x = happydf.Healthy_life_expectancy
y = happydf.Score

xp = np.linspace(-0.02, 1.2, 156)
pr = np.polyval(data_polreg, xp)

fig = mpl.figure(figsize = (14,7))
ax = fig.add_subplot()
ax.grid()

ax.set_xlabel("Healthy life")
ax.set_ylabel("Score")
ax.set_title("Healthy life expectancy vs Score")

ax.scatter(x,y, color = "lightgrey")
ax.plot(xp, pr, color = "red")
```

```python
rel_happy_life = scp.pearsonr(happydf.Score, happydf.Healthy_life_expectancy)
rel_happy_gdp = scp.pearsonr(happydf.Score, happydf.GDP_per_capita)

print(rel_happy_life)
print(rel_happy_gdp)
```

```python
happy_lowdf=happydf[happydf['Score']<4.3]
happy_lowdf 
```

```python
happy_highdf = happydf[happydf['Score']>= 6.5]
happy_highdf
```

```python
tempavg = happydf[happydf['Score']>=4.3]
happy_avgdf = tempavg[tempavg['Score']<6.5]
happy_avgdf
```

```python
vartest_low = np.poly1d(happy_lowdf.Score)
vartest_avg = np.poly1d(happy_avgdf.Score)
vartest_high = np.poly1d(happy_highdf.Score)

stat, p = scp.levene(vartest_low, vartest_avg, vartest_high)
print('l-statistics = ', stat)
print('p value =', p)

alpha = 0.05
if p>alpha:
    print('same variance')
else:
    print('diff variance')
```

```python
krus, p = scp.kruskal(vartest_low, vartest_avg, vartest_high)


print('k-statistics = ', krus)
print('p value =', p)

alpha = 0.05
if p>alpha:
    print('no big diffrence between variables')
else:
    print('big diff')
```
